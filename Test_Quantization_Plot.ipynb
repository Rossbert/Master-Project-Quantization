```python
import numpy as np
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
import pandas as pd
from typing import List
# plt.style.use('bmh')
df = pd.read_csv('Performance.csv')
PLT_SAVE_PATH = './images/'
```

```python
ax : List[List[plt.Axes]]
fig, ax = plt.subplots(2,2, figsize = (12,10))
fig.suptitle('Q Aware model accuraccy degradation with 1 bit affected')

ax[0,0].yaxis.set_major_formatter(ticker.PercentFormatter())
ax[0,0].plot(df.loc[df['kernel_index'] == 0]['bit_disrupted'], df.loc[df['kernel_index'] == 0]['q_aware_acc_degradation']*100, '.')
ax[0,0].set_title('First convolutional layer')
ax[0,0].set_xlabel('Bit affected')
ax[0,0].grid()

ax[0,1].yaxis.set_major_formatter(ticker.PercentFormatter())
ax[0,1].plot(df.loc[df['kernel_index'] == 1]['bit_disrupted'], df.loc[df['kernel_index'] == 1]['q_aware_acc_degradation']*100, '.')
ax[0,1].set_title('Second convolutional layer')
ax[0,1].set_xlabel('Bit affected')
ax[0,1].grid()

ax[1,0].yaxis.set_major_formatter(ticker.PercentFormatter())
ax[1,0].plot(df.loc[df['kernel_index'] == 2]['bit_disrupted'], df.loc[df['kernel_index'] == 2]['q_aware_acc_degradation']*100, '.')
ax[1,0].set_title('Third convolutional layer')
ax[1,0].set_xlabel('Bit affected')
ax[1,0].grid()

ax[1,1].yaxis.set_major_formatter(ticker.PercentFormatter())
ax[1,1].plot(df.loc[df['kernel_index'] == 3]['bit_disrupted'], df.loc[df['kernel_index'] == 3]['q_aware_acc_degradation']*100, '.')
ax[1,1].set_title('Fully connected dense layer')
ax[1,1].set_xlabel('Bit affected')
ax[1,1].grid()
plt.savefig(PLT_SAVE_PATH + 'Q_Aware_acc_degradation.png')
plt.show()
plt.close()
```
